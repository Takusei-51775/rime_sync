## Genshin out in rime format

```python
from pypinyin import lazy_pinyin

word_dict = {}
with open("./dicts/scraper_intermediates/raw/genshin.yaml", 'r', encoding="UTF-8") as infile:
    lines = infile.readlines()
    for line in lines:
        line = line.strip()
        entries = line.split('\t')
        if len(entries) == 0:
            continue
        word = entries[0]

        if len(entries) == 2:
            pron = entries[1]
            word_dict[word] = pron
        else:
            if word in word_dict:
                continue
            else:
                word_dict[word] = None

for word, pron in word_dict.items():
    if pron is None:
        word_dict[word] = ' '.join(lazy_pinyin(word))

words = list(word_dict.items())
words.sort(key = lambda x: str(len(x[0])) + ''.join(x[1].split()))

with open("./dicts/scraper_intermediates/raw/genshin_out.yaml", 'w', encoding="UTF-8") as outfile:
    for entry in words:
        word, pron = entry
        outline = word + '\t' + pron + '\n'
        outfile.writelines(outline)
```

```python
def is_chinese(string):
    for ch in string:
        if not u'\u4e00' <= ch <= u'\u9fff':
            return False
    return True


slangs = set()
with open("./dicts/scraper_intermediates/raw/genshin_slang.yaml", 'r', encoding="UTF-8") as infile:
    lines = infile.readlines()
    for line in lines:
        line = line.strip()
        if not is_chinese(line):
            continue
        slangs.add(line)



slangs = list(slangs)
slangs.sort(key = lambda x: len(x))

for word in slangs:
    print(word)
```
